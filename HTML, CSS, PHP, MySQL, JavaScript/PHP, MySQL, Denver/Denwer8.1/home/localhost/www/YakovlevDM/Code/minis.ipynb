```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi

# Функция для 1.14 (периодическое расширение)
def f_114(t):
    T = 8
    t_mod = t % T  # Упрощенный расчет t_mod
    if 4 < t_mod < 8:
        return 0.5 * t_mod - 2
    elif 0 <= t_mod <= 4:
        return -t_mod + 4
    else:
        return np.nan

# Функция для S3(t) для 1.14
def S3_114(t):
    inst = pi*t/4
    S3 = 3/2
    S3 += 12/pi**2*cos(inst)+2/pi*sin(inst)
    S3 += 1/pi*sin(inst*2)
    S3 += 4/(3*pi**2)*cos(inst*3)+2/(3*pi)*sin(inst*3)
    return S3

# Функция для S6(t) для 1.14
def S6_114(t):
    inst = pi*t/4
    S6 = 3/2
    S6 += 12/pi**2*cos(inst)+2/pi*sin(inst)
    S6 += 1/pi*sin(inst*2)
    S6 += 4/(3*pi**2)*cos(inst*3)+2/(3*pi)*sin(inst*3)
    S6 += 1/(2*pi)*sin(inst*4)
    S6 += 12/(25*pi**2)*cos(inst*5)+2/(5*pi)*sin(inst*5)
    S6 += 1/(3*pi)*sin(inst*6)
    return S6

# Функция для 3.14 (периодическое расширение)
def f_314(t):
    T = 4
    t_mod = t % T  # Упрощенный расчет t_mod
    if 2 < t_mod < 4:
        return 3 * t_mod - 10
    elif 0 <= t_mod <= 2:
        return -t_mod + 2
    else:
       return np.nan

# Функция для S3(t) для 3.14
def S3_314(t):
    inst = pi*t/2
    S3 = 0
    S3 += 8/(pi**2)*cos(inst)+4/pi*sin(inst)
    S3 += 2/(pi**2)*cos(inst*2)-2/pi*sin(inst*2)
    S3 += 8/(9*pi**2)*cos(inst*3)+4/(3*pi)*sin(inst*3)
    return S3

# Функция для S6(t) для 3.14
def S6_314(t):
    inst = pi*t/2
    S6 = 0
    S6 += 8/(pi**2)*cos(inst)+4/pi*sin(inst)
    S6 += 2/(pi**2)*cos(inst*2)-2/pi*sin(inst*2)
    S6 += 8/(9*pi**2)*cos(inst*3)+4/(3*pi)*sin(inst*3)
    S6 += 1/(2*pi**2)*cos(inst*4)-1/pi*sin(inst*4)
    S6 += 8/(25*pi**2)*cos(inst*5)+4/(5*pi)*sin(inst*5)
    S6 += 2/(9*pi**2)*cos(inst*6)-2/(3*pi)*sin(inst*6)
    return S6

# Графики для 1.14
t_114 = np.linspace(-8, 16, 400)
f_values_114 = [f_114(t) for t in t_114]
S3_values_114 = [S3_114(t) for t in t_114]
S6_values_114 = [S6_114(t) for t in t_114]

plt.figure(figsize=(10, 6))
plt.plot(t_114, f_values_114, label='f(t)')
plt.plot(t_114, S3_values_114, label='S3(t)')
plt.plot(t_114, S6_values_114, label='S6(t)')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Графики f(t), S3(t), S6(t) для 1.14 (Периодическое расширение)')
plt.legend()
plt.grid(True)
plt.show()

# Графики для 3.14
t_314 = np.linspace(-4, 8, 400)
f_values_314 = [f_314(t) for t in t_314]
S3_values_314 = [S3_314(t) for t in t_314]
S6_values_314 = [S6_314(t) for t in t_314]

plt.figure(figsize=(10, 6))
plt.plot(t_314, f_values_314, label='f(t)')
plt.plot(t_314, S3_values_314, label='S3(t)')
plt.plot(t_314, S6_values_314, label='S6(t)')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Графики f(t), S3(t), S6(t) для 3.14 (Периодическое расширение)')
plt.legend()
plt.grid(True)
plt.show()
```
